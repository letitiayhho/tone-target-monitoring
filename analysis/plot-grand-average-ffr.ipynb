```python
files = glob.glob('../data/bids/derivatives/preprocessing/sub-*/*fif.gz')
epochs = []
conditions = ['11', '12', '13', '21', '22', '23', '31', '32', '33']
all_evokeds = []

for condition in conditions:
    for file in files:
        sub_epochs = mne.read_epochs(file)

        if not epochs: # handle first loop
            epochs = sub_epochs
            continue

        if sub_epochs.info['nchan'] != 62: # handle files with incorrect number of channels
            print(f"Skipping {file}, incorrect number of channels.")
            continue

        print(f"Compute evokeds for {file} for condition {condition}")
        evokeds = epochs[condition].average()
        all_evokeds.append(evokeds)

    print(f"Combine evokeds for {condition}")
    combined_evokeds = mne.combine_evoked(all_evokeds, "equal")

    print(f"Saving to evokeds_{condition}.fif.gz")
    combined_evokeds.save(f'../data/bids/derivatives/grand-average-ffr/evokeds_{condition}.fif.gz', overwrite = True)
```
